```python
import pandas as pd
```

```python
dfRent = pd.read_csv('/content/bigdata_madrid_idealista_rent.csv')
dfBuy = pd.read_csv('/content/bigdata_madrid_idealista_sales.csv').drop(columns=['highlight'])
df = pd.concat([dfRent, dfBuy], ignore_index=True)
df = df[["latitude", "longitude"]]
df.head()
```

```python
df.plot.scatter(x="latitude", y="longitude")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


kmeans = KMeans(n_clusters=20, random_state=0, n_init="auto").fit(df)
centroids = kmeans.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]

plt.scatter(df["latitude"], df["longitude"], c=kmeans.labels_.astype(float))
plt.scatter(centroids_x, centroids_y, marker = "x", c="red")
plt.show()
```
